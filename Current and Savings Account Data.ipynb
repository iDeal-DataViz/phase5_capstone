# Cleaning & Processing

# Filter Out Rows Without Application ID

```python
currentsavings_df = currentsavings_df[currentsavings_df['Application ID'].notna()]
```

# Convert Date Fields to Datetime Format
### Important for temporal filtering, aggregation, or trend analysis.

```python
date_cols = ['Open Date', 'Closed Date', 'Last Transaction Date', 'SpshotMonth']
for col in date_cols:
    currentsavings_df[col] = pd.to_datetime(currentsavings_df[col], errors='coerce')
```

# Aggregate at Application ID Level
# Because each row is a monthly snapshot, we need to reduce to one row per applicant before merging.

```python
# List of helpful numeric features
agg_features = {
    'Average Balance': ['mean', 'max', 'min'],
    'Overdraft Limit': ['max'],
    'Number of Days in Debit during Month': ['sum', 'mean'],
    'Number of Days in Credit during Month': ['sum', 'mean'],
    'Sum of Monthly Debit Transactions': ['mean'],
    'Sum of Monthly Credit Transactions': ['mean'],
    'Number of ATM Withdrawals': ['sum'],
    'Number of Bounced Cheques': ['sum'],
    'Number of Joint Holders': ['max']
}

# Group by Application ID
banking_summary = currentsavings_df.groupby('Application ID').agg(agg_features)

# Flatten column names
banking_summary.columns = ['_'.join(col).strip() for col in banking_summary.columns.values]
banking_summary.reset_index(inplace=True)
```

## Handle High-Missing Columns
### Fields like:

### Sum of Bounced Cheque transactions (~99.9% missing)

### Sum of Cheque Debit Transactions, POS, Over the Counter Withdrawals



```python
# Drop extremely sparse columns
drop_cols = [
    'Sum of Bounced Cheque transactions',
    'Sum of Cheque Debit Transactions',
    'Sum of Debit Card POS Transactions',
    'Sum of Over the Counter Withdrawals'
]
currentsavings_df.drop(columns=drop_cols, inplace=True)
```

# Feature Engineering Ideas.
### These can improve your model's understanding of financial health:

```python
# Has bounced cheques?
currentsavings_df['Has_Bounced_Cheques'] = (currentsavings_df['Number of Bounced Cheques'] > 0).astype(int)

# Net cash flow for the month
currentsavings_df['Net_Cashflow'] = (
    currentsavings_df['Sum of Monthly Credit Transactions'] -
    currentsavings_df['Sum of Monthly Debit Transactions']
)

# Liquidity ratio: max balance to average balance
currentsavings_df['Liquidity_Ratio'] = (
    currentsavings_df['Maximum End of Day Balance'] /
    currentsavings_df['Average Balance'].replace(0, np.nan)
)
```

```python
currentsavings_df.to_csv("clean_current_savings.csv", index = False)
```
