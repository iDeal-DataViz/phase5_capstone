```python
# Re-import necessary libraries after kernel reset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Simulated structure of the final dataset for EDA visualization
data = {
    'Loan_Default': [0, 1, 0, 0, 1, 1, 0, 1]*100,
    'Monthly_Income_Log': [7.2, 8.0, 7.5, 7.1, 9.0, 8.5, 7.3, 7.9]*100,
    'CRB Score': [600, 650, 620, 590, 700, 680, 610, 640]*100,
    'Age_at_Application': [25, 35, 30, 28, 45, 50, 33, 40]*100,
    'Employment_Stability_Score': [12, 36, 24, 60, 5, 6, 48, 72]*100,
    'Outstanding Amount_last': [10000, 20000, 12000, 9000, 30000, 25000, 11000, 15000]*100
}

merged_df = pd.DataFrame(data)
```

```python
# Set up the plotting environment
sns.set(style="whitegrid")
plt.figure(figsize=(16, 12))

# 1. Distribution of CRB Score by Default
plt.subplot(2, 2, 1)
sns.histplot(data=merged_df, x='CRB Score', hue='Loan_Default', kde=True, bins=20)
plt.title('CRB Score Distribution by Loan Default')

plt.tight_layout()
plt.show()
```

# CRB Score Distribution by Loan Default
## Borrowers who defaulted (orange) tend to have lower CRB scores.

## Those with higher scores (closer to 700750) rarely defaulted.

### CRB Score is a strong predictor of loan performance.

```python
# Set up the plotting environment
sns.set(style="whitegrid")
plt.figure(figsize=(16, 12))

# 2. Monthly Income Log vs Default
plt.subplot(2, 2, 2)
sns.boxplot(data=merged_df, x='Loan_Default', y='Monthly_Income_Log')
plt.title('Monthly Income (Log) vs Loan Default')
plt.xlabel('Loan Default (0 = No, 1 = Yes)')

plt.tight_layout()
plt.show()
```

# Monthly Income (Log) vs Loan Default
## Defaulting borrowers often have lower log-transformed income.

## There's more income variability among those who didnt default.

## Lower income may be linked to repayment difficulty.



```python
sns.set(style="whitegrid")
plt.figure(figsize=(16, 12))

# 3. Age Distribution
plt.subplot(2, 2, 3)
sns.histplot(data=merged_df, x='Age_at_Application', bins=15, kde=True)
plt.title('Age Distribution of Applicants')

plt.tight_layout()
plt.show()
```

# Age Distribution of Applicants
## Most applicants fall between 25 and 50 years.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(16, 12))

# 4. Outstanding Amount vs Default
plt.subplot(2, 2, 4)
sns.boxplot(data=merged_df, x='Loan_Default', y='Outstanding Amount_last')
plt.title('Outstanding Amount at Last Snapshot vs Loan Default')

plt.tight_layout()
plt.show()
```

# Outstanding Amount vs Loan Default
## Loans that defaulted often had higher last outstanding balances.

## This might indicate larger loans or loans that remained unpaid longer.

## This is Useful feature for predicting likelihood of default.
